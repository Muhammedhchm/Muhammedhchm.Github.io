# Example 2 : Multible minimia & Advanced Function

## $$g(x) =X^4-4X^2+5$$

```python
X_2 = (np.linspace(-2,2 ))
```

```python
# defining the g(x) function and dg(x)
def g(x) : 
    return x**4 - 4 * x **2 + 5

def dg(x) :
    return 4*x**3 - 8 * x +5
```

```python
plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,2,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X_2,g(X_2))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('g(X)'  , fontsize = 14)
plt.xlim(-2 , 2 )



    
# chart 2 : derivative 

plt.subplot(1,2,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('dg(X)'  , fontsize = 14)
plt.plot(X_2,dg(X_2))




9 

plt.show()
```

```python
                                                # Gradient Descent 2
def gradient_descent(derivative_func , initial_guess , multipier =0.01 , precision = 0.0001 ) :
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multipier * gradient
    
        step_size = abs(new_x - previous_x)
    
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
    
    
    
        
        if step_size < precision :
            print('Loop ran this many time :', n)
            break

    return new_x , x_list , slope_list

```

```python
local_min , list_x , deriv_list = gradient_descent(dg , 0.5 , 0.02 , 0.001)
print ('local min occurs at : ' , local_min)
print ('num of steps : ' , len(list_x))
```

```python
local_min , list_x , deriv_list = gradient_descent(derivative_func = dg , initial_guess = 2)
print ('local min occurs at : ' , local_min)
print ('num of steps : ' , len(list_x))
```
