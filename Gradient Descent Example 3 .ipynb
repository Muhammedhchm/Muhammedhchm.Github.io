# Example 3 - Divergance 
## $$h(X) = X^5 - 2X^4 +2$$

```python
import matplotlib.pyplot  as plt
import numpy as np
#making data 
X3 = np.linspace(-2.5 , 2.5 , num = 1000)
# defining the g(x) function and dg(x)
def h(x) : 
    return x**5 - 2*x **4 + 2

def dh(x) :
    return 5*x**4 - 8 * x**3 
```

```python
plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,2,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X3,h(X3))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('h(X)'  , fontsize = 14)
plt.xlim(-2 , 2 )



    
# chart 2 : derivative 

plt.subplot(1,2,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('dh(X3)'  , fontsize = 14)
plt.plot(X3,dh(X3))




9 

plt.show()
```

```python
                                                # Gradient Descent 2
def gradient_descent(derivative_func , initial_guess , multipier =0.01 , precision = 0.0001 ) :
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multipier * gradient
    
        step_size = abs(new_x - previous_x)
    
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
    
    
    
        
        if step_size < precision :
            print('Loop ran this many time :', n)
            break

    return new_x , x_list , slope_list
```

```python
local_min , list_x , deriv_list = gradient_descent(derivative_func = dh , initial_guess = 0.2)


plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,2,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X3,h(X3))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('h(X3)'  , fontsize = 14)
plt.xlim(-2 , 2 )

plt.scatter(list_x, h(np.array(list_x)) ,color = 'red' )



    
# chart 2 : derivative 

plt.subplot(1,2,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('dh(X3)'  , fontsize = 14)
plt.plot(X3,dh(X3))

plt.scatter(list_x, deriv_list , color = 'red' )

 

plt.show()

print ('local min occurs at : ' , local_min)
print ('cost at the min occurs at : ' , h(local_min))
print ('num of steps : ' , len(list_x))
```
