# Example (1)

### $$f(x) = x^2 + x +1 $$

```python

def f(x) :
    return x**2 + x + 1
```

```python
X1 = np.linspace (start=-3 , stop = 3 , num = 10)
X1
```

```python
plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,2,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X1,f(X1))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('F(X)'  , fontsize = 14)
plt.ylim(0,12)
plt.xlim(-3 , 3 )

# chart 2 : derivative 

plt.subplot(1,2,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('DF(X)'  , fontsize = 14)
plt.plot(X1,df(X1))
plt.show()

```

```python
def df(x) : 
    return 2*x + 1
```

## Python Loops & Gradient Descent 

```python
                                                # For Loop
for n in range (5) :
    print('hello world'  , n )
print('End of Loops')
```

```python
                                                # while Loop
n =0
while n < 7 :
    print ('Counting ... ' , n)
    n = n + 1
```

```python
                                                # Gradient Descent 
new_x = 3
previous_x = 0
step_multipier = 0.1

for n in range(30):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multipier * gradient

print ('local minimum occurs at :' , new_x)     # = 0.5
print ('the slope or DF(X) value at this point is :' , df(new_x))   # = 0
print ('F(X) value or cost at this point is :' , f(new_x))
```

```python
                                                # Gradient Descent 2
new_x = 3
previous_x = 0
step_multipier = 0.1
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(41):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multipier * gradient

    step_size = abs(new_x - previous_x)
    print(step_size)

    x_list.append(new_x)
    slope_list.append(df(new_x))



    
    if step_size < precision :
        print('Loop ran this many time :', n)
        break
    


print ('local minimum occurs at :' , new_x)     # = 0.5
print ('the slope or DF(X) value at this point is :' , df(new_x))   # = 0
print ('F(X) value or cost at this point is :' , f(new_x))
```

```python
plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,2,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X1,f(X1))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('F(X)'  , fontsize = 14)
plt.ylim(0,12)
plt.xlim(-3 , 3 )


values = np.array(x_list)
plt.scatter(x_list , f(values)  , color='red')
    
# chart 2 : derivative 

plt.subplot(1,2,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('DF(X)'  , fontsize = 14)
plt.plot(X1,df(X1))

plt.scatter(x_list , slope_list , color='red')





plt.show()
```

```python
plt.figure(figsize = [15,5])

# chart 1 : Function 


plt.subplot(1,3,1)


plt.title('Cost function'  ,  fontsize = 17)
plt.plot(X1,f(X1))
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('F(X)'  , fontsize = 14)
plt.ylim(0,12)
plt.xlim(-3 , 3 )


values = np.array(x_list)
plt.scatter(x_list , f(values)  , color='red')
    
# chart 2 : derivative 

plt.subplot(1,3,2)
plt.title('slope of the cost function'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('DF(X)'  , fontsize = 14)
plt.plot(X1,df(X1))

plt.scatter(x_list , slope_list , color='red')

# chart 3 : derivative (close up )

plt.subplot(1,3,3)
plt.title('Gradient Descent close up'  ,  fontsize = 17)
plt.xlabel('X'  ,  fontsize = 14)
plt.ylabel('DF(X)'  , fontsize = 14)

plt.plot(X1,df(X1) , color = 'blue' , linewidth = 6 , alpha = 0.8)

plt.scatter(x_list , slope_list , color='red' , s= 300)


```
